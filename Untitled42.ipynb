```
import seaborn as sns # For loading datasets
import pandas as pd # manipulation
```

```
# Load tips
df = sns.load_dataset("tips")
df.sample(5)
```

# [ILOC](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html)

- Purely integer-location based indexing for selection by position.
.iloc[] is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array.

```
# Indexing a single row
# Wrap your value in a list to return a Pandas DF
df.iloc[[23]]
```

```
# Indexing specific column from a DF using a list
even_ls = list(range(2,12,2)) # Making an even number list

df.iloc[even_ls]
```

```
# Index by list slicing - [start:stop:step]
df.iloc[2:12:2]
```

```
# Index using a boolean array (condition)
df.iloc[list(df["total_bill"] > 50)]
```

# [LOC](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html)

- Access a group of rows and columns by label(s) or a boolean array.
.loc[] is primarily label based, but may also be used with a boolean array.

```
df2 = sns.load_dataset("titanic")
df2.sample(5)
```

```
# Indexing a single value and only certain columns
df2.loc[[23]]
```

```
# Indexing based on a condition
df2.loc[df2["age"] > 70]
```

```
# Only grabbing certain columns with the condition
df2.loc[df2["age"] > 70][["survived", "age", "fare", "who"]]
```

```
# Grab all rows but only certain columns, only look at first 5 rows
df2.loc[:,["survived", "sex", "alive", "who"]].head()
```
